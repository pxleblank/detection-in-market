```python
# Создание видео с предсказанными лейблами для теста
from ultralytics import YOLO
import cv2
import time

model = YOLO("runs/final_train/weights/best.pt")

VIDEO_PATH = "data/train.mp4"
cap = cv2.VideoCapture(VIDEO_PATH)

# Подготовка выходного видео
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
out = cv2.VideoWriter('out_test_train_with_FPS.mp4',
                      cv2.VideoWriter.fourcc(*'mp4v'),
                      fps,
                      (frame_width, frame_height))


prev_time = 0
while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break
    
    current_time = time.time()
    
    results = model.predict(frame, conf=0.2)

    annotated_frame = results[0].plot()
    
    fps_yolo = 1 / (current_time - prev_time) if prev_time != 0 else 0
    prev_time = current_time

    # Отображение FPS на кадре
    cv2.putText(annotated_frame, f"YOLO FPS: {fps_yolo:.2f}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    out.write(annotated_frame)

cap.release()
out.release()
```
